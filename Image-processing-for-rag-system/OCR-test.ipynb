# OCR
In this notebook we test a number of images that contain the Persian language.

first lets initialize the files path.

```python
file_paths = {
    0: "bilboard-mahris.png",
    1: "bilboard-snapp.webp",
    2: "bilboard-valiasr.jpeg",
    3: "book-trip.webp",
    4: "magazine-cars.jpeg",
    5: "magazine-danestaniha.jpeg",
    6: "pixel.png",
}
```

```python
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

model = ocr_predictor(pretrained=True)
def extract_texts(image_path):
    img_doc = DocumentFile.from_images(image_path)
    return model(img_doc)
```

```python
extracted_texts = {}

for key, value in file_paths.items():
    extracted_texts[key] = extract_texts("data/" + value).render()
```

```python
from IPython.display import Image, display

for key, value in extracted_texts.items():
    if "webp" in  file_paths[key]:
        continue
    display(Image("data/" + file_paths[key]))
    print(value)
```
